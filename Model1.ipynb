# Let's Build a System that recommends based on Popularity

<p>
    <b>For this we will require</b>

            1. Dataset which has the ratings of every book
            2. Sorting the book in descending order of their ratings
</p>
    
    

```python
df_books_with_rating = df_books.merge(df_ratings,on='ISBN')
```

```python
df_books_with_rating
```

### Now let's have a look at the ratings each book have, we can do this by grouping the books and then counting the number of ratings they have

```python
df_only_ratings = df_books_with_rating.groupby('Book-Title').count()['Book-Rating'].reset_index() #if we do not use reset index we will not be getting a dataframe
```

<p>
    Now we have a problem, the problem is that inorder to get the book with the highest rating we need to sort the book-rating
    column, but this won't give the right prediction, so it's better to find the average rating of each book and then sort it
    , which can be done using the mean function.
</p>

```python
df_average = df_books_with_rating.groupby('Book-Title').mean()['Book-Rating'].reset_index()
df_average.rename(columns={'Book-Rating':'Average-Rating'},inplace=True)
```

```python
df_average.head()
```

```python
df_popular = df_only_ratings.merge(df_average,on='Book-Title')
```

```python
df_popular
```

```python
df_popular = df_popular[df_popular['Book-Rating']>=250].sort_values('Average-Rating',ascending=False).head(50)
```

```python
df_popular = df_popular.merge(df_books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','Book-Rating','Average-Rating']]
```

# Collaborative Filtering Based Recommender System

<p>
    Recommending similar books
</p>


```python
"""
Arrange Books based on user ratings
"""
active_users = df_books_with_rating.groupby('User-ID').count()['Book-Rating']>200
active_users[active_users]
```

```python
active_users = active_users[active_users].index
```

```python
filtered_rating = df_books_with_rating[df_books_with_rating['User-ID'].isin(active_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = df_books[df_books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('You Belong To Me')
```

```python
pickle.dump(pt,open('pt.pkl','wb'))
pickle.dump(df_books,open('books.pkl','wb'))
pickle.dump(similarity_scores,open('similarity_scores.pkl','wb'))
```
